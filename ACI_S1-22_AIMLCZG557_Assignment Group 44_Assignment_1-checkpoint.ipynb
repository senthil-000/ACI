```python
#Code Block : Set Initial State (Must handle dynamic inputs)
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from memory_profiler import profile
from queue import PriorityQueue

ROW=10
COL=10

# To store Matrix Co-ordinated
class Point:
    def __init__ (self,x:int,y:int):
        self.x=x
        self.y=y
        
# Node Structure
class QueueNode:
    def __init__ (self,pt:Point,dist:float,pri:float):
        self.pt=pt
        self.dist=dist
        self.pri=pri
```

```python
#Code Block : Write function to design the Transition Model/Successor function. Ideally this would be called while search algorithms are implemented
#Traversal seq
rowNum=[1,0,0,-1]
colNum=[0,1,-1,0]
def isValid(row:int,col:int):
    return(row >=0 ) and (row < ROW) and (col >=0 ) and (col < COL)
```

```python
#Code block : Write fucntion to handle goal test (Must handle dynamic inputs). Ideally this would be called while search algorithms are implemented
def goalTest(pt:Point,dest:Point):
    if pt.x == dest.x and pt.y == dest.y:
        return True 
```

### 2.	Definition of Algorithm 1 (Mention the Name of the algorithm here eg., INFORMED SEARCH 

```python
#Code Block : Function for algorithm 1 implementation

def mineField(mat,src:Point,dest:Point):
    
    visited = [[False for i in range(COL)] for j in range(ROW)]
    
    visited[src.x][src.y] = True

    q = PriorityQueue()
    s = QueueNode(src,0.0,1.0)
    q.put((0,0.0,s))

#Minefield traversal
    while q:
        
        if q.empty():
            
            vi = pd.DataFrame(visited) 
            fig, (ax1, ax2) = plt.subplots(ncols=2)
            fig.subplots_adjust(wspace=0.01)
            sns.heatmap(mat,annot=True,linewidth= .5,ax=ax1)
            
            sns.heatmap(visited,annot=True,linewidth= .5,ax=ax2)
            ax2.yaxis.tick_right() 
            plt.show()
            return -1
        
#get next best node for traversal
        temp  = q.get()
        curr = temp[2]
        pt = curr.pt
     
        if not visited[pt.x][pt.y]:
            print("Goal Test on Node ","("+str(pt.x)+","+str(pt.y)+")")
         
      
        if (goalTest(pt,dest)):
            visited[pt.x][pt.y]= -1
            print("        ++++++++ Path exists to Destination ===> ","(",pt.x,",",pt.y,")"," ++++++++")
            vi = pd.DataFrame(visited) 
            fig, (ax1, ax2) = plt.subplots(ncols=2)
            fig.subplots_adjust(wspace=0.01)
            sns.heatmap(mat,annot=True,linewidth= .5,ax=ax1)
            
            sns.heatmap(visited,annot=True,linewidth= .5,ax=ax2)
            ax2.yaxis.tick_right() 
            plt.show()
            
            return curr
        
        visited[pt.x][pt.y]= curr.pri
        
        while not q.empty():
            q.get()
            
        Nbest_adj = 99.0
        Nbest_row=0
        Nbest_col=0
        Nnum=0
        switch=0
        for i in range(4):
           
            row = pt.x + rowNum[i]
            col = pt.y + colNum[i]
            switch=switch+1 
            if(i==2):
                switch=3
            if (isValid(row,col) and ( mat[row][col] >= 0) and not visited[row][col]):

                    if (mat[row][col] < Nbest_adj):
                        if(i==1):
                            switch=switch-2
                   
                    Nbest_adj=mat[row][col]
                    Nbest_row=row
                    Nbest_col=col
                    Nnum = Nnum+1
                    Adjcell = QueueNode(Point(Nbest_row,Nbest_col),curr.dist+mat[Nbest_row][Nbest_col],mat[Nbest_row][Nbest_col])
                    q.put((switch,mat[Nbest_row][Nbest_col],Adjcell))

    return -1
                        
```
